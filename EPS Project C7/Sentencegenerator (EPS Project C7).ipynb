```python
import itertools

def get_sentences():

    setname      = "Close"

    Candidates_1 = [ "the house",  "the restaurant", "the hardware store", "the campus", "the deli"]

    Candidates_2 = ["the forest", "the local pub", "the mountain range", "grandma's house", "the library"]

    Keywords     = ["walk", "drive"]

    Connectives  = ['but', 'however', 'so', 'therefore']

    

    parameters = [ Candidates_1, Candidates_2, Connectives, Keywords]

    sentences = {}

    limit = 100

    for i, combination in enumerate(itertools.product(*parameters)):

        if limit and i >= limit: break

        key = f"{setname}-{i}"

        can1, can2, connective, keyword = combination

        sentence_words = [ can1, 'will', 'be', 'close', 'and', can2, 'will','be','far', connective, 'I', 'will' , keyword ,'to', '_']

        sentence = " ".join(sentence_words) + "."

        

        if connective in ["because", "so", "therefore", "since"]:

             if keyword in ["drive"]: 

                 sentences[key] = ( sentence,  can2, can1 )

             else:

                 sentences[key] = ( sentence,  can1, can2 )

        else:

             if keyword in ["drive"]: 

                 sentences[key] = (sentence,  can1, can2 )

             else: 

                 sentences[key] = (sentence,  can2, can1 )     

    return sentences



get_sentences()
```
