```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import*
import numpy as np
```

```python
def euler(x, f, h):
    return x + f(x) * h
```

```python
def modified_euler(x, f, h):
    k1 = h * f(x)
    k2 = h * f(x + k1)
    return x + 0.5 * (k1+k2)
```

```python
#Q1
def RK4(x,f,h):
    k1 = h * f(x)
    k2 = h * f(x+(1/2)*k1)
    k3 = h * f(x+(1/2)*k2)
    k4 = h * f(x+k3)
    return x+(1/6)*(k1+2*k2+2*k3+k4)
```

```python
def exp_diffeq(x):
    xprime = x
    return xprime
```

```python
#Q2
x, h = 1, 0.01
for i in range(100):
    x = euler(x, exp_diffeq, h)
    print(x)
```

```python
x, h = 1, 0.01
for i in range(100):
    x = modified_euler(x, exp_diffeq, h)
    print(x)
```

```python
x, h = 1, 0.01
for i in range(100):
    x = RK4(x, exp_diffeq, h)
    print(x)
```

```python
#Each of the integration schemes seem to give a similar approxiamation for e around 2.718 which suggests the integration schemes work.
```

```python
#Q3

plt.figure(figsize = (10,5))
plt.title('Differential Equation Solvers with stepsize h=1')
plt.xlabel('$t$'); plt.ylabel('$x(t)$')

x = np.arange(0,5,10**-6)
y = [e**i for i in x]
plt.plot(x,y, label = 'Exact')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/1):
    x = euler(x,exp_diffeq,1)
    x_values.append(x)
    t_values.append(i*1)
    i+=1
plt.plot(t_values,x_values,label = 'Euler')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/1):
    x = modified_euler(x,exp_diffeq,1)
    x_values.append(x)
    t_values.append(i*1)
    i+=1
plt.plot(t_values,x_values,label = 'Modified Euler')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/1):
    x = RK4(x,exp_diffeq,1)
    x_values.append(x)
    t_values.append(i*1)
    i+=1
plt.plot(t_values,x_values,label = 'Runge-Kutta Method')
plt.legend(loc='best')

plt.figure(figsize = (10,5))
plt.title('Differential Equation Solvers with h=0.1')
plt.xlabel('$t$'); plt.ylabel('$x(t)$')

x = np.arange(0,5,10**-6)
y = [e**i for i in x]
plt.plot(x,y, label = 'Exact')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/0.1):
    x = euler(x,exp_diffeq,0.1)
    x_values.append(x)
    t_values.append(i*0.1)
    i+=1
plt.plot(t_values,x_values,label = 'Euler')


x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/0.1):
    x = modified_euler(x,exp_diffeq,0.1)
    x_values.append(x)
    t_values.append(i*0.1)
    i+=1
plt.plot(t_values,x_values,label = 'Modified Euler')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/0.1):
    x = RK4(x,exp_diffeq,0.1)
    x_values.append(x)
    t_values.append(i*0.1)
    i+=1
plt.plot(t_values,x_values,label = 'Runge-Kutta Method')
plt.legend(loc = 'best')

plt.figure(figsize = (10,5))
plt.title('Differential Equation Solvers with h=0.01')
plt.xlabel('$t$'); plt.ylabel('$x(t)$')

x = np.arange(0,5,10**-6)
y = [e**i for i in x]
plt.plot(x,y, label = 'Exact')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/0.01):
    x = euler(x,exp_diffeq,0.01)
    x_values.append(x)
    t_values.append(i*0.01)
    i+=1
plt.plot(t_values,x_values,label = 'Euler')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/0.01):
    x = modified_euler(x,exp_diffeq,0.01)
    x_values.append(x)
    t_values.append(i*0.01)
    i+=1
plt.plot(t_values,x_values,label = 'Modified Euler')

x_values = [1]
t_values = [0]
x=1
i=1
while i<=(5/0.01):
    x = RK4(x,exp_diffeq,0.01)
    x_values.append(x)
    t_values.append(i*0.01)
    i+=1
plt.plot(t_values,x_values,label = 'Rugne-Kutta Method')
plt.legend(loc = 'best')
```

```python
#The solutions of the differential equation solvers look more and more accurate the smaller the step size gets
#The Runge-Kutta Method appears to be the most accurate which makes sense since it is a fourth order method
#The Euler Method appears to be the least accurate which makes sense since it is a linear method
```

```python
#Q4

plt.figure()
plt.xlabel('step size');plt.ylabel('global error')
plt.title('Error in using the Euler Method')
x_values = []
h_values = [1,0.1,0.01,0.001,0.0001]
i=1
x=1
for h in h_values:
    while i<=(5/h):
        x = euler(x,exp_diffeq,h)
        i+=1
    x_values.append(abs(x-e**5))
    i=1
    x=1 
plt.loglog(h_values,x_values)
#I've used a loglog plot to see the exponential relationship between h and the error

plt.figure()
plt.xlabel('step size');plt.ylabel('global error')
plt.title('Error in using the Modified Euler Method')
x_values = []
h_values = [1,0.1,0.01,0.001,0.0001]
i=1
x=1
for h in h_values:
    while i<=(5/h):
        x = modified_euler(x,exp_diffeq,h)
        i+=1
    x_values.append(abs(x-e**5))
    i=1
    x=1 
plt.loglog(h_values,x_values)
#I've used a loglog plot to see the exponential relationship between h and the error

plt.figure()
plt.xlabel('step size');plt.ylabel('global error')
plt.title('Error in using the 4th Order Runge-Kutta Method')
x_values = []
h_values = [1,0.1,0.01,0.001,0.0001]
i=1
x=1
for h in h_values:
    while i<=(5/h):
        x = RK4(x,exp_diffeq,h)
        i+=1
    x_values.append(abs(x-e**5))
    i=1
    x=1 
plt.loglog(h_values,x_values)
#I've used a loglog plot to see the exponential relationship between h and the error
```

```python
#gradient of Euler Method line is roughly 0.874 which is close to 1, this show that the error scales linearly 
#with h
#gradient of Modified Euler Method line is roughly 1.903 which is close to 2, this shows the error scales quadratically 
#with h
#gradient of Runge Kutta Method line is roughly 3.09 which is close to 4, this shows the error scales with h**4
```

```python
#Question 1
import numpy.linalg as la

def three_body(rv):
    d01 = la.norm(rv[0][0]-rv[1][0])
    d02 = la.norm(rv[0][0]-rv[2][0])
    d03 = la.norm(rv[1][0]-rv[2][0])
    dr0dt = rv[0][1]
    dr1dt = rv[1][1]
    dr2dt = rv[2][1]
    dv0dt = (Gm[1]*(rv[1][0]-rv[0][0])/d01**3)+(Gm[2]*(rv[2][0]-rv[0][0])/d02**3)
    dv1dt = (Gm[0]*(rv[0][0]-rv[1][0])/d01**3)+(Gm[2]*(rv[2][0]-rv[1][0])/d03**3)
    dv2dt = (Gm[1]*(rv[1][0]-rv[2][0])/d03**3)+(Gm[0]*(rv[0][0]-rv[2][0])/d02**3)
    return np.array([[dr0dt, dv0dt], [dr1dt, dv1dt],[dr2dt,dv2dt]])
```

```python
#Question 1

Gm = np.array([1,1,1])
rv = np.array([[[1, 0, 0], [0.347111,0.532728,0]],  [[-1, 0, 0], [0.347111,0.532728,0]],[[0,0,0],[-2*(0.347111),-2*(0.532728),0]]])

t   = 0
r0i = [rv[0][0]]
r1i = [rv[1][0]]
r2i = [rv[2][0]]
v0i = [rv[0][1]]
v1i = [rv[1][1]]
v2i = [rv[2][1]]
ti  = [t]

nstep, h = 63244, 0.0001
for i in range(nstep):
    rv = RK4(rv,three_body,h)
    t += h
    r0i.append(rv[0][0])
    r1i.append(rv[1][0])
    r2i.append(rv[2][0])
    v0i.append(rv[0][1])
    v1i.append(rv[1][1])
    v2i.append(rv[2][1])
    ti.append(t)


R0 = np.stack(r0i)
R1 = np.stack(r1i)
R2 = np.stack(r2i)
V0 = np.stack(v0i)
V1 = np.stack(v1i)
V2 = np.stack(v2i)

plt.plot(R0[:, 0], R0[:, 1],label = 'Planet 1') 
plt.plot(R1[:, 0], R1[:, 1],label = 'Planet 2') 
plt.plot(R2[:, 0], R2[:, 1],label = 'Planet 3')
plt.legend(loc = 'best')
plt.title('Orbits of a three-body system-Figure of 8')
plt.xlabel('$x$'); plt.ylabel('$y$')
plt.show() 
```
