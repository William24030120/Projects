```python
class_names = ['Myocardial infarction', 'Heart failure', 'Bundle branch block', 'Dysrhythmia', 'Myocardial hypertrophy',
               'Valvular heart disease', 'Myocarditis', 'Miscellaneous', 'Healthy control', 'Angina']
```

```python
labels = []
for filename in filenames:
        fields = 0
        X, fields = wfdb.io.rdsamp(filename, sampfrom = 0, sampto = 10000)
        if 'Reason for admission: Myocardial infarction' in fields['comments'][4]:
            labels.append(0)
        if 'Reason for admission: Healthy control' in fields['comments'][4]:
            labels.append(1)
        if 'Reason for admission: Valvular heart disease' in fields['comments'][4]:
            labels.append(2)
        if 'Reason for admission: Dysrhythmia' in fields['comments'][4]:
            labels.append(3)
        if 'Reason for admission: Heart failure (NYHA 3)' in fields['comments'][4] or 'Reason for admission: Heart failure (NYHA 2)' in fields['comments'][4] or 'Reason for admission: Heart failure (NYHA 4)' in fields['comments'][4]:
            labels.append(4)
        if 'Reason for admission: Cardiomyopathy' in fields['comments'][4]:
            labels.append(5)
        if 'Reason for admission: Palpitation' in fields['comments'][4]:
            labels.append(6)
        if 'Reason for admission: Hypertrophy' in fields['comments'][4]:
            labels.append(7)
        if 'Reason for admission: Bundle branch block' in fields['comments'][4]:
            labels.append(8)
        if 'Reason for admission: n/a' in fields['comments'][4]:
            labels.append(9)
        if 'Reason for admission: Unstable angina' in fields['comments'][4] or 'Reason for admission: Stable angina' in fields['comments'][4]:
            labels.append(10)
        if 'Reason for admission: Myocarditis' in fields['comments'][4]:
            labels.append(11)
labels = np.array(labels)
```

```python
sample_test_data = []
for x in testing_data:
    idx = np.random.randint(10000,size = 500)
    sample_test_data.append(x[idx,:])
sample_test_data = np.array(sample_test_data)

sample_train_data = []
for x in training_data:
    idx = np.random.randint(10000,size = 500)
    sample_train_data.append(x[idx,:])
sample_train_data = np.array(sample_train_data)
```

```python
input_img = tf.keras.Input(shape=(None,None,500))

x = tf.keras.layers.Conv1D(,100, activation = 'relu', padding = 'same', input_shape = input_img[1:])(input_img)
x = tf.keras.layers.MaxPooling1D(pool_size = 4, strides = 1, padding = 'same')(x)
x = tf.keras.layers.Conv1D(20,10, activation = 'relu', padding = 'same')(x)
x = tf.keras.layers.MaxPooling1D(pool_size = 4, strides = 1, padding = 'same')(x)
x = tf.keras.layers.Conv1D(4, 5, activation = 'relu', padding = 'same')(x)
encoded = tf.keras.layers.MaxPooling1D(pool_size = 4, strides = 1, padding = 'same')(x)


x = tf.keras.layers.Conv1D(4, 5, activation = 'relu', padding = 'same')(encoded)
x = tf.keras.layers.UpSampling1D(5)(x)
x = tf.keras.layers.Conv1D(20,10, activation='relu', padding='same')(x)
decoded = tf.keras.layers.Reshape(500,)

autoencoder = tf.keras.Model(input_img, decoded)
autoencoder.summary()
```

```python
sample_train_data = sample_train_data.reshape((len(sample_train_data), np.prod(sample_train_data.shape[1:])))
sample_test_data = sample_test_data.reshape((len(sample_test_data), np.prod(sample_test_data.shape[1:])))
autoencoder = tf.keras.Model(input_img, decoded)
autoencoder.compile(optimizer ='adam',loss= tf.keras.losses.MeanSquaredError(), metrics = ['accuracy'])
autoencoder.fit(sample_train_data, sample_train_data, epochs = 20,shuffle = True, validation_data = (sample_test_data, sample_test_data))
```
