```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import InputLayer, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from pandas import *
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler


data = pd.read_csv('cover_data.csv')
forest_data = pd.DataFrame(data = data )
forest_data_features = forest_data.iloc[:,0:10]
forest_data_labels = forest_data.iloc[:,-1]
forest_data_features = pd.get_dummies(forest_data_features)
ct  = ColumnTransformer([('only numeric', StandardScaler(),['Elevation','Aspect','Slope','Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology','Horizontal_Distance_To_Roadways','Hillshade_9am','Hillshade_Noon', 'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points'])])
le = LabelEncoder()



forest_train_data_features, forest_test_data_features, forest_train_data_labels, forest_test_data_labels = train_test_split(forest_data_features, forest_data_labels, test_size = 0.2, random_state = 22)
forest_train_data_features = ct.fit_transform(forest_train_data_features)
forest_test_data_features = ct.transform(forest_test_data_features)
forest_train_data_labels = le.fit_transform(forest_train_data_labels)
forest_test_data_labels = le.transform(forest_test_data_labels)


forest_train_data_labels = to_categorical(forest_train_data_labels, dtype = 'int32', num_classes = 7)
forest_test_data_labels = to_categorical(forest_test_data_labels, dtype = 'int32', num_classes = 7)

print(forest_test_data_features.shape)
print(forest_test_data_labels.shape)

model = Sequential()
model.add(InputLayer(input_shape = forest_train_data_features.shape[1]))
model.add(Dense(100, activation = 'relu'))
model.add(Dense(50, activation = 'relu'))
model.add(Dense(7, activation = 'softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#model.fit(forest_train_data_features, forest_train_data_labels, epochs = 50, batch_size = 500, verbose = 1)
#loss, acc = model.evaluate(forest_test_data_features, forest_test_data_labels)
#print(loss, acc)
```
